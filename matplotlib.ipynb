```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xpoints=np.array([1,2])
ypoints=np.array([5,10])

plt.plot(xpoints,ypoints)
plt.show()
```

```python
xpoints=np.array([1,20])
ypoints=np.array([10,200])
plt.plot(xpoints,ypoints,"*")
plt.show()
```

```python
ypoints=np.array([3,8,1,10])

plt.plot(ypoints,'*''b')
plt.show()
```

```python
ypoints=np.array([3,8,1,12])
xpoints=np.array([5,6,7,12])
plt.plot(xpoints,ypoints,marker='o',ms=50,color='red')
plt.show()
```

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='o',ms=20,mfc='r',mec='k')
plt.show()
```

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='o',ms=200,mec='y',mfc='k')
plt.show()
```

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,linestyle='dashed',marker='o',mec='hotpink',ms=20,mfc='blue',linewidth=1)
plt.show()
```

```python
plt.plot(ypoints,linestyle='dotted',marker="o",ms=25,mfc="green",mec="red")
```

```python
xpoints=np.array([10,20,30])
ypoints=np.array([1,2,6])
plt.plot(xpoints,ypoints,marker="*",color="red",mfc="g",mec="r",ms=20,linestyle="dashed")
```

```python
ypoints=np.array([3,8,1,10])
xpoints=np.array([2,4,6,8])
plt.plot(xpoints,ypoints,linewidth=1,linestyle="dotted",color="red",marker="o",ms="20",mec="g")
plt.show()
```

```python
y1=np.array([3,8,1,10])
y2=np.array([6,2,5,1,10,12])
plt.plot(y1)
plt.plot(y2)
plt.show()
```

```python
x=np.array([80,85,90,10,105,110,115,120,125])
y=np.array([240,250,270,280,290,300,310,320,330])

plt.plot(x,y,color="r",linewidth=1)

plt.xlabel("average pulse",fontsize=10,color="red")
plt.ylabel("calories burnage",fontsize=10,color="g")
plt.title("sample",fontsize=50,color="blue")
plt.plot(x,y)
plt.show()
```

```python
x=np.array(["a","b","c","d"])
y=np.array([3,8,1,10])
plt.bar(x,y,color="black",width=0.2)
plt.show()
```

```python
y=np.array([10,135,25,15,19])
mylabels=["ram","gita","apple","banana","fruit"]
plt.pie(y,labels=mylabels)
plt.show()
```
